```python
import pandas as pd
import numpy as np

header=pd.MultiIndex.from_product([['Semester1','Semester2'],['Maths','Science']])
d=([[12,45,67,56],[78,89,45,67],[45,67,89,90],[67,44,56,55]])
df3=pd.DataFrame(d,
                index=['Alisa','Bobby','Cathrine','Jack'],
                columns=header)
df3                                                                         #جدول های 3 لایه مثل pivot table
```

```python
#stack the data frame
stacked_df3=df3.stack()
stacked_df3
```

```python
#unstack the dataframe
unstacked_df3=stacked_df3.unstack()
unstacked_df3
```
